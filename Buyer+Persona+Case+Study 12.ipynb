```python
ax = sns.countplot(x = 'Gender',data = df) 

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Cost'].sum().sort_values(by='Cost', ascending=False)

sns.barplot(x = 'Gender',y= 'Cost' ,data = sales_gen)
```

```python
ax = sns.countplot(data = df,x = 'Age', hue ='Gender')

for bar in ax.cointainers:
    ax.bar_label(bars) 
```

```python
sales_age = df.groupby(['Age'], as_index=False)['Cost'].sum().sort_values(by='Cost', ascending=False)

sns.barplot(x = 'Age',y= 'Cost' ,data = sales_age)
```

```python
df.columns
```

```python
sales_state = df.groupby(['State'], as_index=False)['Gross Sales'].sum().sort_values(by='Gross Sales', ascending=False).head(10)

sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Gross Sales')
```

```python
sales_state = df.groupby(['State'], as_index=False)['Discounts'].sum().sort_values(by='Discounts', ascending=False).head(10)

sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(data = sales_state, x = 'State',y = 'Discounts')
```

```python
df.columns
```

```python
ax = sns.countplot(data = df, x = 'Category')

sns.set(rc={'figure.figsize':(37,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Category'], as_index=False)['Cost'].sum().sort_values(by='Cost', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Category',y= 'Cost')
```

```python
# top 10 most sold products (same thing as above)

fig1, ax1 = plt.subplots(figsize=(10,7))
df.groupby('Date')['Gross Sales'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
